8.How do you apply a custom function to create a new column that categorizes candies into â€˜High Sugarâ€™ or â€˜Low Sugarâ€™ based on the â€˜sugarpercentâ€™?

```python
def catgeory(sugar):
    if sugar>0.5:
        return "High_sugar"
    else :
        return "Low_sugar"
df["value"]= df["sugarpercent"].apply(catgeory)
df[["competitorname","value","sugarpercent"]].head(10)
```
